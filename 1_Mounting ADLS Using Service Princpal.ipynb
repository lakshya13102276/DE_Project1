Mounting ADLS on Azure Databricks for easy access and secure access

```
client_id = dbutils.secrets.get(scope="saprojectone-scope", key = "saprojectone-service-principal-clientid")
tenant_id = dbutils.secrets.get(scope="saprojectone-scope", key = "saprojectone-service-principal-tenentid")
client_secret = dbutils.secrets.get(scope="saprojectone-scope", key = "saprojectone-service-principal-clientsecret")
```

```
configs = {"fs.azure.account.auth.type": "OAuth",
          "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
          "fs.azure.account.oauth2.client.id": client_id,
          "fs.azure.account.oauth2.client.secret": client_secret,
          "fs.azure.account.oauth2.client.endpoint": f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"}

```

```
# Optionally, you can add <directory-name> to the source URI of your mount point.
dbutils.fs.mount(
  source = "abfss://bronze-project1@saprojectone.dfs.core.windows.net/",
  mount_point = "/mnt/saprojectone/bronze-project1",
  extra_configs = configs)
```
